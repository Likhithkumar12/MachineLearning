```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('bhp.csv')
df.head()
```

```python
df.describe()
```

```python
%matplotlib inline
plt.hist(df.price_per_sqft,bins=20,rwidth=0.6)
plt.xlabel('price_per_sqft')
plt.ylabel('count')
plt.yscale('log')
plt.show()
```

```python
mini,maxi=df['price_per_sqft'].quantile([0.001,0.999])
df[(df['price_per_sqft']<mini)|(df['price_per_sqft']>maxi)]
```

```python
df2=df[(df['price_per_sqft']>mini)&(df['price_per_sqft']<maxi)]
df2
```

```python
df.shape[0]-df2.shape[0]
```

```python
max_limit=df2.price_per_sqft.mean()+4*df2.price_per_sqft.std()
min_limit=df2.price_per_sqft.mean()-4*df2.price_per_sqft.std()
df2[(df2['price_per_sqft']<min_limit)|(df2['price_per_sqft']>max_limit)]
```

```python
df3=df2[(df2['price_per_sqft']>min_limit)&(df2['price_per_sqft']<max_limit)]
df3
```

```python
df2.shape[0]-df3.shape[0]
```

```python
plt.hist(df3.price_per_sqft,bins=20,rwidth=0.6)
plt.xlabel('price_per_sqft')
plt.ylabel('count')

plt.show()
```

```python
import numpy as np
from scipy.stats import norm
plt.hist(df3.price_per_sqft,bins=20,rwidth=0.6,density=True)
plt.xlabel('price_per_sqft')
plt.ylabel('count')
rg=np.arange(df3.price_per_sqft.min(),df3.price_per_sqft.max(),100)
plt.plot(rg,norm.pdf(rg,df3.price_per_sqft.mean(),df3.price_per_sqft.std()))
plt.show()
```

```python

df2['z_score']=(df2.price_per_sqft-df2.price_per_sqft.mean())/(df2.price_per_sqft.std())
df2
```

```python
df5=df2[(df2['z_score']>-4)&(df2['z_score']<4)]
df5
```

```python
df2.shape[0]-df5.shape[0]
```
