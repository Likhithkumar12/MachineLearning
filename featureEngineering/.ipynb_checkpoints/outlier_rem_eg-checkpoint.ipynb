```python
import pandas as pd
```

```python
df=pd.read_csv('heights.csv')
df
```

```python
df.shape
```

```python
df.describe()
```

```python
max=df['height'].quantile(0.95)
max
```

```python
min=df['height'].quantile(0.05)
min
```

```python
df[df['height']>max]
```

```python
df[df['height']<min]
```

```python
df[(df['height']<max) & (df['height']>min)]
```
