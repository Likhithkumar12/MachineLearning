```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
iris=load_iris()
dir(iris)
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
df['target']=iris.target
df
```

```python
df['flower']=df['target'].apply(lambda x:iris.target_names[x])
df
```

```python
%matplotlib inline
df0=df[df.target==0]
df1=df[df.target==1]
df2=df[df.target==2]
```

```python
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='red',marker='*')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='blue',marker='*')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color='green',marker='*')
```

```python
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='red',marker='*')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='blue',marker='*')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='green',marker='*')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(iris.data,iris.target,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=3)
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
y_pred=model.predict(x_test)
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
import seaborn as sn
sn.heatmap(cm,annot=True)
plt.xlabel('predicted')
plt.ylabel('Truth')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
import pandas as pd
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
```

```python
digits=load_digits()
```

```python
dir(digits)
```

```python
df=pd.DataFrame(digits.data,digits.target)
df
```

```python
df['target']=digits.target
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(digits.data,digits.target,test_size=0.2,random_state=10)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn=KNeighborsClassifier(n_neighbors=3)
```

```python
kn.fit(x_train,y_train)
```

```python
kn.score(x_test,y_test)
```

```python
kn=KNeighborsClassifier(n_neighbors=5)
kn.fit(x_train,y_train)
```

```python
kn.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
y_pred=kn.predict(x_test)
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
import seaborn as sn
sn.heatmap(cm,annot=True)
plt.xlabel('predicted')
plt.ylabel('truth')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
