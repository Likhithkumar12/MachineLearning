```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import load_iris
```

```python
df=pd.read_csv("income.csv")
df
```

```python
%matplotlib inline
plt.scatter(df.Age,df['Income($)'])
```

```python
km=KMeans(n_clusters=3)
km
```

```python
df['cluster']=km.fit_predict(df[['Age','Income($)']])
df.head()
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
```

```python
plt.scatter(df1.Age,df1['Income($)'],color='red',marker='*')
plt.scatter(df2.Age,df2['Income($)'],color='green',marker='*')
plt.scatter(df3.Age,df3['Income($)'],color='black',marker='*')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
mm=MinMaxScaler()
mm.fit(df[['Income($)']])
df['Income($)']=mm.transform(df[['Income($)']])
df
```

```python
mm.fit(df[['Age']])
df['Age']=mm.transform(df[['Age']])
df
```

```python
km=KMeans(n_clusters=3)
km
```

```python
df['cluster']=km.fit_predict(df[['Age','Income($)']])
df
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]

plt.scatter(df1.Age,df1['Income($)'],color='red',marker='*')
plt.scatter(df2.Age,df2['Income($)'],color='green',marker='*')
plt.scatter(df3.Age,df3['Income($)'],color='black',marker='*')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
sse=[]
k_range=range(1,10)
for k in k_range:
    km=KMeans(n_clusters=k)
    km.fit(df[['Age','Income($)']])
    sse.append(km.inertia_)
sse
```

```python
plt.plot(k_range,sse,color='yellow',marker='+')
```

```python
iris=load_iris()
dir(iris)
```

```python
iris.feature_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
df=df.drop(['sepal length (cm)','sepal width (cm)'],axis='columns')
df
```

```python
plt.scatter(df['petal length (cm)'],df['petal width (cm)'],color='red',marker='*')
```

```python
km=KMeans(n_clusters=3)
df['clusters']=km.fit_predict(df)
df
```

```python
df1=df[df.clusters==0]
df2=df[df.clusters==1]
df3=df[df.clusters==2]
```

```python
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='red',marker='*')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='green',marker='*')
plt.scatter(df3['petal length (cm)'],df3['petal width (cm)'],color='blue',marker='*')
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.legend()
```

```python
sse=[]
rg=range(1,10)
for k in rg:
    km=KMeans(n_clusters=k)
    km.fit_predict(df)
    sse.append(km.inertia_)
sse
```

```python
plt.plot(rg,sse,color='red',marker='*')
```
